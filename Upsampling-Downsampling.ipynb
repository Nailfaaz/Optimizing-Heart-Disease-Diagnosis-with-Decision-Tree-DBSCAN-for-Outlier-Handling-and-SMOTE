# Upsampling Downsampling

```python
import pandas as pd
from imblearn.over_sampling import SMOTE

# Read the dataset
df = pd.read_csv('data-project-ml-binned.csv')

# Separate features and target variable
X = df.drop(columns=['HeartDiseaseorAttack'])
y = df['HeartDiseaseorAttack']

# Initialize SMOTE
smote = SMOTE(random_state=42)

# Apply SMOTE to up-sample the minority class
X_smote, y_smote = smote.fit_resample(X, y)

# Create a new DataFrame with the up-sampled data
upsampled_df = pd.concat([pd.DataFrame(X_smote), pd.DataFrame(y_smote, columns=['HeartDiseaseorAttack'])], axis=1)

# Save the up-sampled data to a CSV file
# upsampled_df.to_csv('upsampled_data_smote.csv', index=False)
```
